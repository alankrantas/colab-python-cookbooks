## 在 Colab 使用中文字型繪圖的最簡易辦法

### 下載/上傳 OTF 或 TTF 字型到工作階段

以下字型 (思源黑體) 出自 https://github.com/googlefonts/noto-cjk/tree/main/Sans/OTF/TraditionalChinese

```python
!wget 'https://github.com/googlefonts/noto-cjk/raw/main/Sans/OTF/TraditionalChinese/NotoSansCJKtc-Regular.otf'
```

### 將字型檔新增到 matplotlib

```python
from matplotlib import font_manager as fm
fm.fontManager.addfont('./NotoSansCJKtc-Regular.otf')
```

### 從系統字型清單尋找新字型名稱

```python
for font in fm.fontManager.ttflist:
  print(font.name)
```

你也可以用 HTML 繪出系統字型 (不支援中文的只會使用瀏覽器預設中文字型)

```python
from IPython.core.display import HTML

html = []
for font in fm.fontManager.ttflist:
    html.append(f"<p>[{font.name}]<br /><span style='font-family:{font.name}; font-size: 24px;'>matplotlib 字型測試</p>")

HTML(''.join(html))
```

### 匯入套件和指定 matplotlib 字型

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Noto Sans CJK TC'  # 指定 matplotlib 字型
plt.rcParams['font.size'] = 14  # 設定 matplotlib 全域字型大小
```

### 用中文繪製圖表

```python
import numpy as np

x = np.linspace(0, np.pi*2, 100)

plt.figure(figsize=(12, 8))
plt.title('圖表標題', fontsize=28)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.xlabel('橫軸 X', fontsize=18)
plt.ylabel('縱軸 Y', fontsize=18)
plt.xticks(ticks=[0, np.pi*0.5, np.pi, np.pi*1.5, np.pi*2], 
           labels=['π*零', 'π*零點五', 'π*一點零', 'π*一點五', 'π*二點零'], 
           fontsize=12)
plt.legend(labels=['圖例甲', '圖例乙'], 
fontsize=20)
plt.grid(True)

plt.show()
```

對其他使用 matplotlib 的繪圖套件 (pandas, seaborn, scikit-learn 等) 也有作用

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',
                 header=None, 
                 names=('花萼寬度', '花萼長度', '花瓣寬度', '花瓣長度', '花種'))
sns.pairplot(df, hue='花種')
plt.show()
```
